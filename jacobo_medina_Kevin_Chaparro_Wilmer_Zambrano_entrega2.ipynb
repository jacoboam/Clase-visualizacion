###  <center><span style="color:blue">Gráfico datos multivariados - Wilmer Zmbrano</span></center>

```python
tab = df_2019['area_de_conocimiento'].value_counts().rename_axis('label').reset_index(name='counts')
tab
```

```python
import matplotlib.pyplot as plt
from matplotlib import cm
from math import log10

labels = tab['label']
data = tab['counts']

n = len(data)
k = 10 ** int(log10(max(data)))
m = k * (1 + max(data) // k)

r = 1.2
w = r / n 

colors = [cm.gist_ncar(i / n) for i in range(n)]

fig, ax = plt.subplots()
ax.axis("equal")
for i in range(n):
    innerring, _ = ax.pie([m - data[i], 
                           data[i]], 
                          radius = r - i * w, 
                          startangle = 90, 
                          labels = ["", labels[i]], 
                          labeldistance = 1 - 1 / (1.5 * (n - i)), 
                          textprops = {"alpha": 0}, 
                          colors = ["white", colors[i]])
    plt.setp(innerring, 
             width = w, 
             edgecolor = "white")
    radius = r - i * w
    plt.text(0, radius - w / 2, f'{labels[i]} – {data[i]} ', ha='right', va='center')
plt.title('Áreas de Conocimiento Elegidas por los Estudiantes en el Año 2019')
plt.show()
```

+ El gráfico muestra la cantidad de estudiantes admitidos para el año 2019 por área de conocimiento. El gráfico radial Bart chart muestra barras circulares de mayor a menor frecuencia, dependiendo de los estudiantes admitidos por área de conocimiento, por mencionar algunas el área de economía, administración, contaduría y afines con 14221 estudiantes admitidos que fue la categoría con mayor número de registros, por otro lado, la categoría con un menor número de registro fue el área de matemáticas y ciencias naturales con 1341.

### <center><span style="color:black">¿Por qué?</span></center>

+ El objetivo de la visualización es descubrir tendencias entre las áreas de conocimiento más y menos elegidas por los estudiantes en Colombia para el año de 2019. se muestra en el grafico que para la categoría de economía, administración, contaduría y afines el 30.74% de los estudiantes la eligen para estudiar, para ingeniería, arquitectura urbanismo y afines son el 25.49%, para ciencias sociales y humanas cuenta con el 16.57/%, ciencias de la salud el 8.34%, para ciencias de la educación el 8.25%, bellas artes con el 4.43%, agronomía, veterinaria y afines el 3.30% y por ultimo matemáticas y ciencias naturales que es la menos popular entre la población estudiantil ya que cuenta con tan solo un 2.87%.

### <center><span style="color:black">¿Cómo?</span></center>

+ Para la codificación se ordenaron las áreas de conocimiento de mayor a menor frecuencia, realizando el gráfico de de barras radial.

### <center><span style="color:black">Marcas y Canales</span></center>
+ utiliza marca de línea para el atributo cuantitativo.
+ utiliza el canal de color para diferenciar las categorías de áreas de conocimiento.
+ también utiliza el canal de longitud de línea para saber cuáles son las áreas más elegidas por los estudiantes en Colombia.

### <center><span style="color:black">Mejoras posibles</span></center>
+ una mejora para la visualización seria en vez de mostrar la cantidad de estudiantes admitidos por área mostrar los porcentajes para entender más rápido y de una manera más clara los resultados.
