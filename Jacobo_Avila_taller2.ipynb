```python
import pandas as pd
import numpy as np
import seaborn as sns
import networkx as nx
from pyvis.network import Network
import matplotlib.pyplot as plt
import os
```

```python
g = nx.read_gml('football.gml')
print("Número de nodos:", g.number_of_nodes())
print("Número de enlaces:", g.number_of_edges())
```

```python
plt.figure(figsize=(50,40))
plt.subplot(121)
nx.draw(g, with_labels=True,pos=nx.kamada_kawai_layout(g))
```

### ¿Qué?

Es un dataset de redes que nos muestra los partidos entre los equipos universitarios de futbol americano durante la temporada regular en el año 2000 donde tenemos la variable cualitativa que es el nombre de cada equipo y luego esta cada nodo y enlace que hay , es decir contra quien jugo cada equipo en esa temporada.

### ¿Por qué?

Los datos ya venían para usar redes así que no toca hacer ninguna transformación o algún cambio para obtener la visualización, lo que se busca con la visualización principalmente es disfrutar, ya mirando un poco mas a fondo pues sería explorar la topología e identificar como estan conenctados los nodos como para intentar identificar en que conferencia esta cada equipo.

### ¿Cómo?

Por como se encuentra el data set y como la variable cualitativa no tiene jerarquía se utiliza el modismo de redes, y se intentan organizar los enlaces de manera optima para evitar lo que mas se pueda la oclusión.

### Marcas y Canales

- marca de puntos para los nodos
- marca de lineas para los enlaces
- canal de posición para minimizar los cruces

Me parece que al ser para disfrutar generalmente cumple las reglas por que el 2D está justificado ya que lo que mas se busca con el modismo es disfrutar pues al ponerlo en 1D sería muy aburrido, también el modismo es muy sencillo así que el espectador no tiene que matarse por guardar mucha información sino solo observar y buscar, lo único que si es complicado de obtener es que se observe todo fácilmente ya que al haber tantos nodos hay un poco de oclusión pero se hace lo posible para reducir lo que más se pueda.
